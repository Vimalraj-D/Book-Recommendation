```python
import pandas as pd
import re
```

```python
df = pd.read_csv("dataset\\Book_Dataset.csv")

print("Original shape:", df.shape)
```

```python
def clean_price(price):
    if pd.isna(price):
        return None
    price = re.sub(r"[₹,]", "", str(price))
    try:
        return float(price)
    except:
        return None
```

```python
df["Price"] = df["Price"].apply(clean_price)
```

```python
df["Rating"] = pd.to_numeric(df["Rating"], errors="coerce")
df["No. of People rated"] = pd.to_numeric(
    df["No. of People rated"], errors="coerce"
)
```

```python
df = df.drop_duplicates(subset=["Title"])
```

```python
df = df.fillna({
    "Main Genre": "Unknown",
    "Sub Genre": "Unknown",
    "Type": "Unknown",
    "Price": df["Price"].median(),
    "Rating": df["Rating"].mean(),
    "No. of People rated": 0
})
```

```python
print("After cleaning:", df.shape)
```

```python
def create_document(row):
    return f"""
Title: {row['Title']}
Author: {row['Author']}
Main Genre: {row['Main Genre']}
Sub Genre: {row['Sub Genre']}
Type: {row['Type']}
Price: {row['Price']}
Rating: {row['Rating']}
People Rated: {row['No. of People rated']}
"""
```

```python
df["document"] = df.apply(create_document, axis=1)
```

```python
df.to_pickle("dataset\\cleaned_book_dataset.pkl")
```
